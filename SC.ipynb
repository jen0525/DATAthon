```python
import pandas as pd
import numpy as np

# 양키 스타디움 좌표 (위도, 경도)
YANKEE_LAT = 40.8296
YANKEE_LON = -73.9262

# Haversine 거리 계산 함수 (단위: km)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km)
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    a = np.sin(delta_phi / 2.0) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return R * c


# CSV 파일 경로
df_latitude_longitude_csv_path = '/content/drive/My Drive/Colab Notebooks/projects/DATAthon/Airbnb/df_latitude_longitude.csv'

# CSV 파일 로드
df = pd.read_csv(df_latitude_longitude_csv_path, usecols=['id', 'latitude', 'longitude'])  # 필요한 열만 선택


# 거리 계산 및 반경 10km 내 여부 판별
df["distance_km"] = df.apply(lambda row: haversine(YANKEE_LAT, YANKEE_LON, row["latitude"], row["longitude"]), axis=1)
df["within_10km"] = df["distance_km"] <= 10  # 10km 이내 여부

output_path = "/content/drive/My Drive/Colab Notebooks/projects/DATAthon/Airbnb/nyc_listings_within_10km.csv"
df.to_csv(output_path, index=False)
print(f" 파일 저장 완료: {output_path}")

df.head()
```
