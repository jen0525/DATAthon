욕시를 벗어난 숙소라도 야구 경기장과 가까운지 확인

뉴욕시에는 메이저리그 야구 경기장이 두 곳이 있음

Yankee Stadium (뉴욕 양키스)

위도: 40.8296, 경도: -73.9262
Citi Field (뉴욕 메츠)

위도: 40.7571, 경도: -73.8458

```python
import numpy as np

# Haversine 공식을 사용해 두 지점 간 거리(단위: km) 계산 함수
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반경 (km)
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    a = np.sin(delta_phi / 2) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2) ** 2
    return 2 * R * np.arcsin(np.sqrt(a))

# 경기장 좌표
yankee_stadium = (40.8296, -73.9262)
citi_field = (40.7571, -73.8458)

# 이상치 숙소와 야구 경기장 간 거리 계산
outliers['distance_to_yankee'] = outliers.apply(lambda row: haversine(row['latitude'], row['longitude'], *yankee_stadium), axis=1)
outliers['distance_to_citi'] = outliers.apply(lambda row: haversine(row['latitude'], row['longitude'], *citi_field), axis=1)

# 결과 확인 (거리 기준 정렬)
outliers[['id', 'name', 'latitude', 'longitude', 'distance_to_yankee', 'distance_to_citi']].sort_values(by='distance_to_yankee')
```

```python
# 야구장과 10km 초과 거리인 숙소 ID 추출
outlier_ids_to_remove = outliers[outliers['distance_to_yankee'] > 10]['id']

# 원본 데이터에서 해당 ID를 가진 숙소 제거
filtered_df = df[~df['id'].isin(outlier_ids_to_remove)]

# 결과 확인
print(f"제거된 숙소 개수: {len(outlier_ids_to_remove)}")
print(f"제거 후 남은 숙소 개수: {len(filtered_df)}")
```

```python
# 'host_name'과 'name' 결측치는 'Unknown'으로 채우기
df['host_name'].fillna('Unknown', inplace=True)
df['name'].fillna('No Name', inplace=True)

# 'last_review'와 'reviews_per_month' 결측치는 0으로 채우기
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
from geopy.distance import geodesic

# Yankee Stadium과 Citi Field의 위도, 경도
yankee_stadium = (40.8296, -73.9262)  # Yankee Stadium
citi_field = (40.7573, -73.8458)      # Citi Field

# 숙소가 야구 구장 근처 3km 이내에 있는지 판단하는 함수
def is_near_stadium(lat, lon, stadium_coords, radius_km=3):
    return geodesic((lat, lon), stadium_coords).km <= radius_km

# 숙소가 3km 이내에 위치한 경우를 필터링
df['near_baseball_stadium'] = df.apply(lambda row: is_near_stadium(row['latitude'], row['longitude'], yankee_stadium) or
                                         is_near_stadium(row['latitude'], row['longitude'], citi_field), axis=1)

# A그룹과 B그룹으로 나누기
df_A = df[df['near_baseball_stadium'] == True]
df_B = df[df['near_baseball_stadium'] == False]

# 각 그룹을 별도의 파일로 저장
df_A.to_csv('/content/drive/MyDrive/Colab Notebooks/airbnb/df_A.csv', index=False)
df_B.to_csv('/content/drive/MyDrive/Colab Notebooks/airbnb/df_B.csv', index=False)

# 저장 완료 메시지
print("A그룹과 B그룹의 데이터를 각각 'df_A.csv'와 'df_B.csv'로 저장했습니다.")
```

```python
df_A = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/airbnb/df_A.csv')
df_B = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/airbnb/df_B.csv')
```

```python
# 시즌 구분 함수 (야구 시즌: 4~10월, 비시즌: 11~3월)
def get_season(month):
    if month in [4, 5, 6, 7, 8, 9, 10]:  # 야구 시즌: 4~10월
        return 'Baseball Season'
    else:  # 비시즌: 11~3월
        return 'Off Season'
```

```python
# 'last_review' 컬럼을 이용해 시즌 구분을 위한 월 추출
df_A['review_month'] = pd.to_datetime(df_A['last_review'], errors='coerce').dt.month
df_B['review_month'] = pd.to_datetime(df_B['last_review'], errors='coerce').dt.month
```

```python
# 각 데이터에 시즌 구분 적용
df_A['season'] = df_A['review_month'].apply(get_season)
df_B['season'] = df_B['review_month'].apply(get_season)

# A그룹과 B그룹에 대해 시즌별로 가격, 리뷰 갯수, 예약 가능 여부 등 비교
df_A_season = df_A.groupby('season').agg(
    avg_price=('price', 'mean'),
    avg_reviews=('number_of_reviews', 'mean'),
    avg_reviews_per_month=('reviews_per_month', 'mean'),
    avg_minimum_nights=('minimum_nights', 'mean'),
    avg_availability=('availability_365', 'mean')
).reset_index()

df_B_season = df_B.groupby('season').agg(
    avg_price=('price', 'mean'),
    avg_reviews=('number_of_reviews', 'mean'),
    avg_reviews_per_month=('reviews_per_month', 'mean'),
    avg_minimum_nights=('minimum_nights', 'mean'),
    avg_availability=('availability_365', 'mean')
).reset_index()

# A그룹과 B그룹의 차이 계산 (각 그룹의 시즌별 비교)
df_comparison = pd.merge(df_A_season, df_B_season, on='season', suffixes=('_A', '_B'))
df_comparison['price_difference'] = (df_comparison['avg_price_A'] - df_comparison['avg_price_B']) / df_comparison['avg_price_B'] * 100
df_comparison['reviews_difference'] = (df_comparison['avg_reviews_A'] - df_comparison['avg_reviews_B']) / df_comparison['avg_reviews_B'] * 100
df_comparison['reviews_per_month_difference'] = (df_comparison['avg_reviews_per_month_A'] - df_comparison['avg_reviews_per_month_B']) / df_comparison['avg_reviews_per_month_B'] * 100
df_comparison['minimum_nights_difference'] = (df_comparison['avg_minimum_nights_A'] - df_comparison['avg_minimum_nights_B']) / df_comparison['avg_minimum_nights_B'] * 100
df_comparison['availability_difference'] = (df_comparison['avg_availability_A'] - df_comparison['avg_availability_B']) / df_comparison['avg_availability_B'] * 100

# 결과 출력
print(df_comparison)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 비교 데이터 준비
data = {
    'season': ['Baseball Season', 'Off Season'],
    'price_difference': [-34.7, -38.4],
    'reviews_difference': [22.42, 30.13],
    'reviews_per_month_difference': [19.04, 31.18],
    'minimum_nights_difference': [-22.33, -12.09],
    'availability_difference': [8.23, -3.85]
}

df_comparison = pd.DataFrame(data)

# 시각화: 각 시즌에 대한 차이 비교
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# 각 그래프 설정
sns.barplot(x='season', y='price_difference', data=df_comparison, ax=axes[0, 0], palette='coolwarm')
axes[0, 0].set_title('Price Difference (%)')

sns.barplot(x='season', y='reviews_difference', data=df_comparison, ax=axes[0, 1], palette='coolwarm')
axes[0, 1].set_title('Reviews Difference (%)')

sns.barplot(x='season', y='reviews_per_month_difference', data=df_comparison, ax=axes[0, 2], palette='coolwarm')
axes[0, 2].set_title('Reviews per Month Difference (%)')

sns.barplot(x='season', y='minimum_nights_difference', data=df_comparison, ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title('Minimum Nights Difference (%)')

sns.barplot(x='season', y='availability_difference', data=df_comparison, ax=axes[1, 1], palette='coolwarm')
axes[1, 1].set_title('Availability Difference (%)')

# 빈 축 제거
axes[1, 2].axis('off')

# 전체 레이아웃 조정
plt.tight_layout()

# 이미지로 저장
plt.savefig('/content/drive/MyDrive/Colab Notebooks/airbnb/season_comparison.png')

# 시각화 출력
plt.show()
```
